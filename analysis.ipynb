```python
#import libraries
import pandas as pd
import seaborn as sns   
```

```python
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#clean up data
#Only want USA Races, 50k or 50Mi, 2020
#There are both 50km and 50mi
df[df['Event distance/length'] == '50mi']
```

```python
#combine 50km and 50mi using isin
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters together
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
#assign it to a variable
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(5)
```

```python
df2.shape
```

```python
#remove (USA) from Event name
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(5)
```

```python
#clean up the athlete age
df2['athelete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athelete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns: Athelete Club, Athelete Country, Athelete year of birth, Athelete Age Category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(5)
```

```python
# clean up null values
df2.isnull().sum()
```

```python
df2[df2['athelete_age'].isnull()==1]
```

```python
# drop null values
df2 = df2.dropna()
```

```python
# check for duplicates
df2.duplicated().sum()
```

```python
# reset index
df2.reset_index(drop=True)
```

```python
df2.dtypes
```

```python
df2['athelete_age'] = df2['athelete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# rename columns
df2 = df2.rename(columns={'Event name': 'race_name', 
                          'Event dates': 'race_day', 
                          'Event distance/length': 'race_length',
                          'Event number of finishers': 'race_number_of_finishers',
                          'Athlete performance': 'athlete_performance', 
                          'Athlete gender': 'athlete_gender',
                          'Athlete average speed': 'athlete_average_speed', 
                          'Athlete ID': 'athlete_id',
                          'Year of event': 'year'})
```

```python
df2.head(5)
```

```python
# reorder columns
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athelete_age', 'athlete_performance', 'athlete_average_speed', 'year']]
```

```python
df3.head(5)
```

```python
sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3, x = 'race_length', hue='athlete_gender')
```

```python
sns.displot(df3[df3['race_length']=='50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data=df3, x='race_length', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quartile', linewidth=1)
```

```python
sns.lmplot(data=df3, x='athelete_age', y='athlete_average_speed', hue='athlete_gender')
```

```python
# Difference in speed for the 50km, 50mi male to female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age group is the fastest for 50Mi Race
df3.query('race_length == "50mi"').groupby('athelete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False).query('count > 19').head(15)
```

```python
# What age group is the Worst for 50Mi Race
df3.query('race_length == "50mi"').groupby('athelete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=True).query('count > 19').head(15)
```
